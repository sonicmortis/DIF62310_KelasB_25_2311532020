1. Impor library pandas yang nantinya akan digunakan untuk membuat dan memanipulasi dataframe
2. Buat variabel bernama data yang memiliki kolom luas dan harga beserta value-nya
3. Ubah variabel data menjadi dataframe yang disimpan didalam df
4. cukup ketik variabel df saja, maka akan muncul tabel dataframe yang sudah dibuat



```python
import pandas as pd

data = {"luas": [6, 8, 10, 14, 18],
        "harga": [7, 9, 13, 17.5, 18]}

df = pd.DataFrame(data)

df
```

impor library matplotlib untuk menghasilkangrafik dimana luas sebagai sumbu x dan harga sebagai sumbu y

atur title, label x dan y seperti di dalam kode, limit x dan y 25 sehingga skala maksimal di kedua sumbu 25, lalu munculkan grid agar lebih mudah dibaca lalu tampilkan

```python
import matplotlib.pyplot as plt

df.plot(kind="scatter", x="luas", y="harga")
plt.title("Perbandingan Luas Tanah dengan Harga Tanah")
plt.xlabel("Luas Tanah (m persegi)")
plt.ylabel ("Harga Tanah (dollar)")
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

```python
import numpy as np

X = np.array(df['luas'])
y = np.array(df['harga'])

print(f"X Shape: {X}")
print(f"y Shape: {y}")
```

```python
X = X.reshape(-1, 1)

print(f"X Shape: {X}")
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
X_vis = np.array((0, 25)).reshape(-1, 1)
y_vis = model.predict(X_vis)
```

```python
plt.scatter(X, y)
plt.plot(X_vis, y_vis, c="red")

plt.title('perbandingan Luas Tanah dengan Harga Tanah')
plt.xlabel('Luas Tanah (m persegi)')
plt.ylabel('Harga Tanah (dollar)')
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

```python
print(f'intercept: {model.intercept_}')
print(f'slope: {model.coef_}')
```

```python
X_train = np.array([6,8,10,14,18]).reshape(-1, 1)
y_train = np.array([7,9,13,17.5,18])

X_test = np.array([8, 9, 11, 16, 12]).reshape(-1, 1)
y_test = np.array([11, 8.5, 15, 18, 11])
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X_test)
r_squared = r2_score(y_test, y_pred)
print(f'R-squared: {r_squared}')
```
